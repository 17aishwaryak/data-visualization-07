```python
import numpy as np
import matplotlib.pyplot as plt


# Create a linearly spaced array
x = np.linspace(0, 5, 10)

# Create a figure and two subplots
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].plot(x,x**2, x, np.exp(x))
axes[0].set_title("Normal scale")
axes[0].plot(x,x**2, x, np.exp(x))
axes[1].set_yscale("log")
axes[1].set_title("logarithmic scale (y)");
```

```python
x = np.linspace(0, 10, 5)
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(x, x**2, lw=2, label=r'$\alpha$')
ax.plot(x, x**3, lw=2, label=r'$\beta$')
ax.plot(x, x**4, lw=2, label=r'$\gamma$')
ax.plot(x, x**5, lw=2, label=r'$\delta$')
ax.plot(x, x**6, lw=2, label=r'$\epsilon$')
ax.set_xticks(x)
ax.set_xticklabels([r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$', r'$\epsilon$'], fontsize=18)
yticks = [0, 50, 100, 150]
ax.set_yticks(yticks)
ax.set_yticklabels(["$%.1f$" % y for y in yticks], fontsize=18)
ax.legend()
```

```python
fig,axl=plt.subplots()
axl.plot(x,x**2,lw=2, color="blue")
axl.set_ylabel(r"area $(m^2)$",fontsize=28,color="blue")
for label in axl.get_yticklabels():
  label.set_color("blue")
# Create a twin axes
ax2 = axl.twinx()
ax2.plot(x,x**3,lw=2,color="red")
ax2.set_ylabel(r"volume $(m^3)$",fontsize=18,
color="red")
for label in ax2.get_yticklabels():
    label.set_color("red")
```
